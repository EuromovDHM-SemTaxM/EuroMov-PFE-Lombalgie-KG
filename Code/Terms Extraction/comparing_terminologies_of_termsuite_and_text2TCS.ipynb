# Comparaison des deux m√©thodes

```python
list_pairs_2 = list()
for mot1 in termsuite_2:
    for mot2 in multilingual_2:
        list_pairs_2.append((levenshtein.normalized_distance(mot1, mot2), mot1, mot2))
```

```python
with open('list_pairs.json', 'w', encoding="utf-8") as f:
    f.write(json.dumps(list_pairs_2))
```

```python
with open('list_pairs.json') as f:
    list_pairs_2 = json.loads(f.read())
```

```python
len(list_pairs_2)
```

```python
paire_similaire = []
print("Les mots les plus similaires sont: ")
for couple in list_pairs_2:
    seuil = 0.2
    if len(couple[1]) <=4 and len(couple[2]) <= 4:
        seuil = 0.25
    if couple[0] <= seuil:
        print(couple[1] + " et " + couple[2])
        paire_similaire.append((couple[1], couple[2]))
```

### Enregistrement des differentes listes (similaires et uniques)

```python
new_termsuite = [mot for mot in termsuite_2 if mot not in [i[0] for i in paire_similaire]]
with open('termsuite_unique.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(new_termsuite))
len(new_termsuite)
```
