## Preprocessing

```python
from nltk.corpus import stopwords
import spacy
# python3 -m spacy download fr_core_news_md
import string
```

```python
stop = set(stopwords.words('french'))
exclude = set(string.punctuation)
nlp = spacy.load('fr_core_news_md')
```

```python
stop
```

```python
allText_lemma = nlp(allText)
```

```python
for i,token in enumerate(allText_lemma):
    print(token, token.lemma_)
    if i == 10:
        break
```

```python
def clean(text):
    normalized = [token.lemma_ for token in text]
    stop_free = [word for word in normalized if word not in stop]
    punc_free = ' '.join(ch for ch in stop_free if ch not in exclude)
    
    return punc_free.split()
```

```python
words = clean(allText_lemma)
```

```python
words
```

```python
import nltk

words_tags = nltk.pos_tag(words)
words_tags
```
